## 3. **Establecemos la GPU por defecto en caso de estar disponible**

```python
# Configurar para que TensorFlow utilice la GPU por defecto
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Configurar para que TensorFlow asigne memoria dinámicamente
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        # Especificar la GPU por defecto
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Manejar error
        print(e)
```

## 5. **Cálculo de características**

```python
# Calculamos y agregamos la measure AspectRatio
AspectRatio = frijoles['MajorAxisLength'] / frijoles['MinorAxisLength']
frijoles['AspectRatio'] = AspectRatio

# Calculamos y agregamos la measure Roundness
Roundness = np.pi*4*frijoles['Area'] / frijoles['Perimeter']**2
frijoles['Roundness'] = Roundness

# Informacion del dataset actualizado
frijoles.info()
```

### 6.1. *Variable: 'Area'*

#### 6.1.1. **Información general**

```python
divider = "-" * 49

print(f'{divider}\n                Variable: "Area"\n{divider}')

print(f'Cantidad de valores unicos: {frijoles["Area"].nunique()}')
print(f'Tipo de dato: {frijoles["Area"].dtype}')

print(f'{divider}\n{frijoles["Area"].describe()}\n{divider}')
```

#### 6.1.2. **Frecuencia absoluta (por intervalos)**

```python
frec_abs_area_intervalos = pd.cut(x=frijoles['Area'], bins=np.arange(20000,256001,2000), right=False)
frec_abs_area_intervalos = frec_abs_area_intervalos.value_counts().to_frame().reset_index().rename(columns={'Area':'intervalos' ,	'count': 'frec_abs'}).sort_values(by='intervalos').reset_index(drop=True)
frec_abs_area_intervalos
```

#### 6.1.3. **Área vs Cantidad de frijoles (agrupado en intervalos)**

```python
plt.figure(figsize=(16,12))
plt.bar(x=np.arange(0,118,1), height='frec_abs', data=frec_abs_area_intervalos, align='edge', width=1, edgecolor='black')
plt.xlabel(xlabel='Intervalos Área')
plt.xticks(ticks=np.arange(0,119,1), rotation=90, labels=np.arange(20000,256001,2000), fontsize=8)
plt.ylabel(ylabel='Cantidad de frijoles')
plt.yticks(ticks=np.arange(0,901,25))
plt.title(label='Área vs Cantidad de frijoles (agrupado en intervalos)')
plt.show()
```

#### 6.1.4. **Distribución de Área de Frijoles**

```python
plt.figure(figsize=(7,5))
plt.boxplot(x=frijoles['Area'], vert=False)
plt.xticks(ticks=np.arange(20000,260001,20000), rotation=90)
plt.yticks(ticks=[])
plt.title(label='Distribución de Área de Frijoles')
plt.show()
```

#### 6.1.5. **Observaciones**

**Área vs Cantidad de frijoles (agrupado en intervalos):**
* Distribución bastante concentrada hacia la izquierda (valores más chicos). Asimetría hacia la derecha
* Tiene un rango bastante grande.
* Más que seguramente haya valores atípicos a la derecha.

**Distribución de Área de Frijoles:**
* Muchos valores atípicos hacia la derecha.
